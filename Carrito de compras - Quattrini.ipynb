```python
def mostrar_menu(nombre_usuario):
    print("Menú de opciones para", nombre_usuario)

def mostrar_productos(productos): 
    print("Productos disponibles en góndola:")
    for producto, precio in productos.items():
        print(f"{producto}: ${precio}")

def agregar_producto(carrito, producto, precio, cantidad):
    carrito.append((producto, precio, cantidad))
    print(f"{cantidad} {producto}(s) agregado(s) al carrito")

def eliminar_producto(carrito, producto):
    for item in carrito:
        if item[0] == producto:
            carrito.remove(item)
            print(f"{producto} eliminado del carrito")
            return
    print(f"{producto} no encontrado en el carrito")

def ver_lista_compras(carrito):
    if not carrito:
        print("El carrito está vacío")
    else:
        print("Lista de compras:")
        for i, (producto, precio, cantidad) in enumerate(carrito, 1):
            print(f"{i}. {cantidad} {producto}: ${precio * cantidad}")

def finalizar_compra(carrito):
    ver_lista_compras(carrito)
    total = sum(precio * cantidad for _, precio, cantidad in carrito)
    print("\nTotal a pagar: $", total)  # Consider formatting the total as currency
    print("¡Gracias por su compra!")
    return total

def main():
    productos = {
        "Leche": 1200,
        "Galletas": 950,
        "Gaseosa": 2000,
        "Huevos": 3500,
        "Aceite": 2100,
        "Pan": 1350,
        "Azucar": 1150,
        "Sahumerios": 575
    }
    carrito = []

    nombre_usuario = input("Por favor, ingrese su nombre: ")
    print("Bienvenido/a", nombre_usuario)

    while True:
        mostrar_menu(nombre_usuario)
        opcion = input("Ingrese el número de la opción que desea: "
                       "1 Ver productos disponibles, "
                       "2 Agregar producto, "
                       "3 Eliminar producto, "
                       "4 Ver lista de compras, "
                       "5 Finalizar compra, "
                       "6 Salir ")

        if opcion == "1":
            mostrar_productos(productos)
            input("Presione enter para visualizar las siguientes opciones")
        elif opcion == "2":
            producto = input("Ingrese el nombre del producto que desea agregar: ")
            if producto in productos:
                cantidad = int(input("Ingrese la cantidad que desea agregar: "))
                agregar_producto(carrito, producto, productos[producto], cantidad)
                input("Presione enter para visualizar las siguientes opciones")
            else:
                print("Producto no válido")
        elif opcion == "3":
            ver_lista_compras(carrito)
            producto = input("Ingrese el nombre del producto que desea eliminar: ")
            eliminar_producto(carrito, producto)
            input("Presione enter para visualizar las siguientes opciones")
        elif opcion == "4":
            ver_lista_compras(carrito)
            input("Presione enter para visualizar las siguientes opciones")
        elif opcion == "5":
            total = finalizar_compra(carrito)
            input("Presione enter para visualizar las siguientes opciones")
            break
        elif opcion == "6":
            print("¡Gracias, vuelva pronto! Buena semana.")
            break
        else:
            print("Opción no válida")

if __name__ == "__main__":
    main()
```
