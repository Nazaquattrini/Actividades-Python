```python
class Tamagotchi:
    def __init__(self, nombre):
        self.nombre = nombre
        self.nivel_energia = 100
        self.nivel_hambre = 0
        self.nivel_felicidad = 50
        self.humor = "indiferente"
        self.esta_vivo = True

    def mostrar_estado(self):
        self.actualizar_humor()
        estado = f"""
        Nombre: {self.nombre}
        Nivel de energía: {self.nivel_energia}
        Nivel de hambre: {self.nivel_hambre}
        Nivel de felicidad: {self.nivel_felicidad}
        Humor: {self.humor}
        """
        print(estado)

    def alimentar(self):
        if self.esta_vivo:
            self.nivel_hambre -= 10
            self.nivel_energia -= 15
            self.verificar_estado()
        else:
            print(f"{self.nombre} está muerto y no puede ser alimentado.")
        self.mostrar_estado()

    def jugar(self):
        if self.esta_vivo:
            self.nivel_felicidad += 20
            self.nivel_energia -= 18
            self.nivel_hambre += 10
            self.verificar_estado()
        else:
            print(f"{self.nombre} está muerto y no puede jugar.")
        self.mostrar_estado()

    def dormir(self):
        if self.esta_vivo:
            self.nivel_energia += 40
            self.nivel_hambre += 5
            self.verificar_estado()
        else:
            print(f"{self.nombre} está muerto y no puede dormir.")
        self.mostrar_estado()

    def verificar_estado(self):
        if self.nivel_hambre >= 20:
            self.nivel_energia -= 20
            self.nivel_felicidad -= 30
            print(f"{self.nombre} tiene demasiada hambre y está perdiendo energía y felicidad.")
        if self.nivel_energia <= 0:
            self.esta_vivo = False
            self.nivel_energia = 0
            print(f"{self.nombre} ha muerto.")
        self.nivel_hambre = max(self.nivel_hambre, 0)
        self.nivel_felicidad = max(self.nivel_felicidad, 0)
        self.nivel_energia = max(self.nivel_energia, 0)

    def actualizar_humor(self):
        if self.nivel_felicidad >= 80:
            self.humor = "eufórico"
        elif self.nivel_felicidad >= 60:
            self.humor = "feliz"
        elif self.nivel_felicidad >= 40:
            self.humor = "indiferente"
        elif self.nivel_felicidad >= 20:
            self.humor = "triste"
        else:
            self.humor = "enojado"

# Función principal para interactuar con el usuario
def main():
    nombre = input("Elige un nombre para tu Tamagotchi: ")
    mi_tamagotchi = Tamagotchi(nombre)
    
    while mi_tamagotchi.esta_vivo:
        print("\n¿Qué te gustaría hacer?")
        print("1. Mostrar estado")
        print("2. Alimentar")
        print("3. Jugar")
        print("4. Dormir")
        print("5. Salir")
        
        opcion = input("Elige una opción (1-5): ")
        
        if opcion == "1":
            mi_tamagotchi.mostrar_estado()
        elif opcion == "2":
            mi_tamagotchi.alimentar()
        elif opcion == "3":
            mi_tamagotchi.jugar()
        elif opcion == "4":
            mi_tamagotchi.dormir()
        elif opcion == "5":
            print("¡Adiós!")
            break
        else:
            print("Opción no válida. Por favor elige una opción del 1 al 5.")
        
        if not mi_tamagotchi.esta_vivo:
            print(f"{mi_tamagotchi.nombre} ha muerto. Fin del juego.")

if __name__ == "__main__":
    main()
```
